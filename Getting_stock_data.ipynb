### Python for FInance: Get Stock Market Data

we begin by using pandas_datareader to import financial stock data https://pydata.github.io/pandas-datareader/index.html


```python
!pip install pandas-datareader
```

```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
```

Step 1 : Specify data range for analysis
Here we begin by creating start and end dates using pythons datatime module 


```python
end = dt.datetime.now()
start = dt.datetime(2000, 1,1)
start, end
```

Select the stocks/tickers you would like to analyse, use the searcch bar in yahoo finance to work out other ticker symbols


```python
stocks = ['AAPL', 'META', 'AMZN', 'XOM']
```

Step 3 call the pandas_datareader datareader module:

two ways of doing this

1. pdr.DataReader(stocks, 'yahoo', start, end)
2. pdr.get_data_yahoo(stocks, start, end)

```python
df = pdr.get_data_yahoo(stocks, start, end)
df.head()
```

Step 4 Understanding the pandas dataframe structure, a good way to quickly understand how a pandas dataframe is structured is to look at the index and the columns 

```python
df.index
```

```python
df.columns
```

Step 5: access useful attributes easily

```python
Close = df.Close
Close.head()
```

Step 6 Gain Insights quickly with pandas describe method

```python
Close[Close.index > end - dt.timedelta(days=100)].describe(percentiles = [0.1, 0.5, 0.9])
```

Step 7: Plotting with matplotlib vs plotly

```python
Close.plot(figsize = (12, 8))
```
